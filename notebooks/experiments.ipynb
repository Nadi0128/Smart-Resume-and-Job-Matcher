```python
import sys
sys.path.append("..")

from app.resume_parser import extract_resume_text
from app.job_parser import parse_job_description
from app.matcher import compute_similarity

# Load resume
with open("../data/sample_resumes/test_cv.pdf", "rb") as f:
    resume_text = extract_resume_text(f)

# Job description example
job_text = """
VINCI Construction SI propose un stage PFE de 5 à 6 mois dès janvier 2026 au poste de chef.fe de projet DATA/Low Code 
au sein de l’équipe Data & AI. Le stagiaire participera à la gestion de projets (réunions, coordination, 
suivi des risques), à l’analyse et visualisation de données, ainsi qu’à la veille et expérimentation de 
nouvelles technologies IA/DATA/Low Code. Le profil recherché est un étudiant Bac+5/6 curieux, autonome,
communicant, avec une première expérience en gestion de projet, un anglais courant et idéalement des 
connaissances en SQL, Power BI, IA/LLM, API ou Azure.
"""

job_text = parse_job_description(job_text)

score = compute_similarity(resume_text, job_text)
print("Matching score:", round(score, 3))
```
